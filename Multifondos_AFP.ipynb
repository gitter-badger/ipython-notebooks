```python
from datetime import datetime
import pandas as pd
import pandas.io.data as web
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
pd.set_option('max_columns', 50)
%matplotlib inline
```

```python
run_console = False
update_from_yahoo = False
update_from_Quandl = False
update_from_dropbox = False
```

```python
if run_console:
    %qtconsole
```

```python
if update_from_yahoo:
    today = datetime.now()
    ipsa = web.DataReader('^IPSA', 'yahoo', start=datetime(2003, 1, 10), end=today)['Close']
    spy = web.DataReader('SPY', 'yahoo', start=datetime(1993, 1, 29), end=today)['Close']
    eem = web.DataReader('EEM', 'yahoo', start=datetime(2003, 4, 15), end=today)['Close']
    vea = web.DataReader('VEA', 'yahoo', start=datetime(2007, 7, 26), end=today)['Close']
    df = pd.DataFrame()
    df['ipsa'] = ipsa
    df['spy'] = spy
    df['eem'] = eem
    df['vea'] = vea
    df.to_csv('INDEX.csv')
```

```python
if update_from_Quandl:
    import Quandl
    df = Quandl.get('BNP/USDCLP')
    df.to_csv('USDCLP.csv')
```

```python
if update_from_dropbox:
    import urllib2
    url = 'http://dl.dropboxusercontent.com/u/35775157/multifondos_afp.txt'
    res = urllib2.Request(url)
    csvio = urllib2.urlopen(res)
    df = pd.read_csv(csvio, index_col=['<TICKER>', '<DTYYMMDD>'], parse_dates=True, usecols=['<TICKER>', '<DTYYMMDD>','<CLOSE>'])
    df = df.unstack('<TICKER>')
    df.columns = df.columns.droplevel()
    df.to_csv('AFP.csv')
```

```python
afp = pd.read_csv('INDEX.csv', parse_dates=True, index_col='Date')
usd = pd.read_csv('USDCLP.csv', parse_dates=True, index_col='Date')
afp['usd'] = usd
```

```python
afp.info()
afp.tail()
```

```python
afp.plot(subplots=True, figsize=(10, 10), sharex=True)
#plt.legend(loc='best')
```

```python
afp_pc = afp.pct_change()
afp_pc.plot(subplots=True, figsize=(10, 10), sharex=True)
```

```python
afp['A'] =  0.20*afp_pc['spy'] + 0.15*afp_pc['ipsa'] + 0.17*afp_pc['eem'] + 0.06*afp_pc['vea'] + 0.50*afp_pc['usd']
afp['A'] = afp['A'].cumsum()
```
